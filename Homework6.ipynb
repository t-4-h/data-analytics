### Practical Question 1

In the first question we will construct a k-means clustering model. We will use the data of food consumption in different countries in Europe.

a) Start by importing the data from the 'EuropeConsumption.csv' file

```python
#Import the data into the 'data' variable below
data = pd.read_csv('EuropeConsumption.csv')

#Check if the data is correctly imported
print(data)
```

b) Create a scatter plot of 'ReadMeat' vs. 'WhiteMeat'. Do not forget to format the graph: include the axes labels and a title for the graph.

```python
#Create the plot here. Do not forget to include the required lines of code to visualize 
#the graph directly in the notebook.

plt.scatter(data.RedMeat, data.WhiteMeat)
plt.title('Red Meat v White Meat')
plt.xlabel('red meat')
plt.ylabel('white meat') 
```

Do you think there is a possible group or cluster relation based on the White and Read meat consumption of different countries? Explain your answer

**Your Answer:** Yes I think there's a possible cluster relation although it is difficult to tell how the data will be clustered since the data points are not clearly segregated from each other as groups. 

c) Create a K-means clustering model with k=3 and fit the data of 'ReadMet' into this model

```python
# Construct your k-means clustering model named 'clustering_model':
clustering_model =  KMeans(n_clusters = 3)      #The model should have three clusters

# Fit the data to the model
# Hint:Reshaping the data into the correct format to fit the model and storing in the 'x' variable
x = data.RedMeat.values.reshape(-1,1)

# Fit the array x into the model
clustering_model.fit(x)
```

d) Predict the values for the array 'x' using our clustering model

```python
# Predict the values of 'x' using the created k-means model and save it into the variable 'predicted_values'
predicted_values = clustering_model.predict(x)
```

e) Visualize your results. Create a scatter plot of RedMeat vs WhiteMeat. In the plot the color of the points should match the group or cluster assigned to them in our prediction section.

```python
# Scatter plot RedMeat vs WhiteMeat where the color of the points matches their cluster or group. 
# Do not forget to format the graph

x = data['RedMeat']
y = data['WhiteMeat']
plt.scatter(x,y,c=predicted_values)
```

Is our clustering model working? Explain your answer

**Your Answer:**  Yes the clustering model is working because it is grouping the data into 3 distinct clusters.

What is this graph telling us? Explain your answer

**Your Answer:** This graph shows 3 clusters grouped by the amount of red meat consumed in each country. 

f) The following lines of code print a scatter plot of RedMeat vs WhiteMeat with different colors for the different clusters. In addition, we can see the names of the countries next to each point.

```python
# Get the RedMeat, WhiteMeat and Country information into variables z,y, and n.
z = data['RedMeat']
y = data['WhiteMeat']
n = data['Country']

# Create a figure for our plot
fig, ax = plt.subplots()
ax.scatter(z, y, c=predicted_values)

# Anotate the names of the countries next to each point
for i, txt in enumerate(n):
    ax.annotate(txt, (z[i], y[i]))

# Format our graph
plt.xlabel('Red Meat')
plt.ylabel('White Meat')
plt.title('Read Meat vs White Meat Consumption')
```

What does this graph tell you about the different countries? Explain your answer

**Your Answer:**  The data is grouped based on red meat consumption and it shows you how much red meat vs white meat a country consumes while allowing you to compare countries to each other based on their groupings and relative locations on the graph.  Some countries consume more white than red meat (and vice versa) while others consume similar amounts (either a lot or a little) of both.  

### Practical Question 2

In this part you should be able to replicate our previous constructed model. However, now you should use 'WhiteMeat' as the parameter for clustering and create two models, one with 3 clusters and another one with 5 clusters.

a) Create a k-means cluster with k=3 and 'WhiteMeat' as our clustering parameter. The final result should be a scatter plot of RedMeat vs. WhiteMeat with the colors of the points reflecting the group or cluster each point belongs to. In addition, the graph should display the name of the corresponding country next to the each point.

```python
## Construct your k-means clustering model named 'clustering_model_1':
clustering_model_1 = KMeans (n_clusters = 3) #The model should have three clusters

# Reshape the data into the correct format to fit the model and save it as 'x'
x = data.WhiteMeat.values.reshape(-1,1)

# Fit the array x into the model
clustering_model_1.fit(x)

# Predict the values of 'x' using the created k-means model and save it into the variable 'predicted_values_1'
predicted_values_1 = clustering_model_1.predict(x)

#Create your scatter plot of RedMeat vs WhiteMeat
#Remember to differentiate the colors of the clusters, to include the names of the countries and to format your graph
y = data['RedMeat']
z = data['WhiteMeat']
n = data['Country']

# Create a figure for our plot
fig, ax = plt.subplots()
ax.scatter(z, y, c=predicted_values)

# Anotate the names of the countries next to each point
for i, txt in enumerate(n):
    ax.annotate(txt, (z[i], y[i]))

# Format our graph
plt.xlabel('Red Meat')
plt.ylabel('White Meat')
plt.title('Read Meat vs White Meat Consumption')
```

What is the main difference between this model and our previous cluster model with RedMeat? Explain your answer

**Your Answer:** Since this model is clustering via white meat consumption, the clusters are grouped along the y axis according to the amount of white meat consumed. Visually, the data is now seperated horizontally rather than vertically. 

b) Create a k-means cluster with k=5 and 'WhiteMeat' as the clustering parameter. The final result should be a scatter plot of RedMeat vs. WhiteMeat with the colors of the points reflecting the group or cluster of the point. In addition the graph should display the name of each country next to the corresponding point.

```python
## Construct your k-means clustering model named 'clustering_model_2':
clustering_model_2 = KMeans(n_clusters = 5) #The model should have five clusters

# Reshaping the data into the correct format to fit the model and save it as 'x'
x = data.WhiteMeat.values.reshape(-1,1)

# Fit the array x into the model
clustering_model_2.fit(x)

# Predict the values of 'x' using the created k-means model and save it into the variable 'predicted_values_2'
predicted_values_2 = clustering_model_2.predict(x)


#Plot your scatter plot of RedMeat vs WhiteMeat
#Remember to differentiate the colors of the clusters and to include the names of the countries
#Do not forget to format your graph
y = data['RedMeat']
z = data['WhiteMeat']
n = data['Country']

# Create a figure for our plot
fig, ax = plt.subplots()
ax.scatter(z, y, c=predicted_values_2)

# Anotate the names of the countries next to each point
for i, txt in enumerate(n):
    ax.annotate(txt, (z[i], y[i]))

# Format our graph
plt.xlabel('Red Meat')
plt.ylabel('White Meat')
plt.title('Red Meat vs White Meat Consumption')
```

What is the main difference between the first model, k=3, and the second model, k=5? Explain your answer

**Your Answer:** The k=5 model seems to be grouped vertically across the x axis rather than horizontally up the y axis.

Comparing the two graphs, k=3 and k=5, what happens to Greece and the UK? Explain your answer.

**Your Answer:** Greece and the UK were grouped differently based on their location due to the increased k value.
